```python
from FlightRadar24 import FlightRadar24API
frAPI = FlightRadar24API()

#Location
center_lat = 29.833753512247522
center_lon = -95.44674947598007
deltaLong = 0.06  #degrees East/West
deltaLat = 0.1 #degrees North

# Define square bounds (min_lat, max_lat, min_lon, max_lon)
min_lat = center_lat 
max_lat = center_lat + deltaLat 
min_lon = center_lon - deltaLong
max_lon = center_lon + deltaLong

bounds = f"{max_lat},{min_lat},{min_lon},{max_lon}"
#bounds = frAPI.get_bounds_by_point(center_lat,center_lon,10000)
#bounds

flights = frAPI.get_flights(bounds = bounds)
flights
```

```python
for flight in flights:
    details = frAPI.get_flight_details(flight)

    callsign = details.get('identification', {}).get('callsign')
    model = details.get('aircraft', {}).get('model', {}).get('text')
    seats = details.get('aircraft', {}).get('capacity', {}).get('seats')

    origin_data = details.get('airport', {}).get('origin', {})
    dest_data = details.get('airport', {}).get('destination', {})

    origin_name = origin_data.get('name')
    origin_city = origin_data.get('position', {}).get('region', {}).get('city')
    origin_country = origin_data.get('position', {}).get('country', {}).get('name')

    dest_name = dest_data.get('name')
    dest_city = dest_data.get('position', {}).get('region', {}).get('city')
    dest_country = dest_data.get('position', {}).get('country', {}).get('name')

    print(f"""
    Callsign: {callsign}
    Seat Capacity: {seats}
    Model: {model}
    Origin: {origin_name} ({origin_city}, {origin_country})
    Destination: {dest_name} ({dest_city}, {dest_country})
    """)
```
